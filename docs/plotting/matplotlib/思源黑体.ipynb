```python
from matplotlib import font_manager
```

`matplotlib`当前版本不支持`ttc`文件格式（font_manager.get_fontext_synonyms('ttf')中没有包含`ttc`），预计3.0版本才支持。解决`ubuntu`操作系统下的绘图问题，使用以下比较笨的方法：
1. 查找本地字体（`fc-list :lang=zh`,一般都会安装思源黑体）
2. 指定字体路径
3. 使用时添加参数`fontproperties`

```python
import matplotlib  
import matplotlib.pyplot as plt 
```

```python
# 注意文件后缀ttc而非ttf
zh_font = font_manager.FontProperties(fname='/usr/share/fonts/opentype/noto/NotoSansCJK-Bold.ttc')
```

```python
# 正负号问题
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
# 之前遭遇过中文正常，而字符不正常情形，故以下坐标均有混搭
fig, ax = plt.subplots()
ax.plot([-1,2,-5,3])
ax.set_title('This is a font: {}'.format('思源黑体(Source Han Sans)/Noto Sans CJK字体介绍'),fontproperties=zh_font)
ax.set_xlabel('x轴123', fontproperties=zh_font)
ax.set_ylabel('y轴123', fontproperties=zh_font)
plt.show()
```
