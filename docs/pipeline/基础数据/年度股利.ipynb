```python
%%zipline --start 2007-1-1 --end 2008-11-24 --capital-base 100000 --bm_symbol 000001

from zipline.pipeline import Pipeline, Fundamentals

from zipline.pipeline.factors import CustomFactor
from zipline.pipeline.data import USEquityPricing
from zipline.api import attach_pipeline, pipeline_output
from zipline.api import symbol, sid, get_datetime

def make_pipeline():
    return Pipeline(
        columns = {
            '股利':Fundamentals.dividend.latest.downsample('year_start'),  
        },
    )

def initialize(context):  
    attach_pipeline(make_pipeline(), 'example')
    context.year = 0
    context.month = 0

def handle_data(context, data):
    today = get_datetime('Asia/Shanghai')
    if today.year != context.year or today.month != context.month:
        output = pipeline_output('example')
        print('\n',today, '\n', output.loc[[sid(2),sid(600645),sid(2025)],:])
        context.year = today.year
        context.month = today.month
```
