```python
%%zipline --start 2018-4-10 --end 2018-4-25
from zipline.api import symbol, order, record, get_current_holdings, get_current_allocations

def initialize(context):
    context.stocks = (symbol(x) for x in ('000333','600771'))

def handle_data(context, data):
    for s in context.stocks:
        order(s, 100)
    his = data[context.stocks]
    record(收盘价=his.price,
           成交量=his.volume,
           总市值=his.tmv)
    print(get_current_holdings(context, data))
    print(get_current_allocations(context, data))
```
